```python
# Coordinates (lon_min, lat_min, lon_max, lat_max)
bounds = {
    "Manhattan": (-74.0479, 40.6829, -73.9067, 40.8820),
    "Brooklyn":  (-74.0419, 40.5701, -73.8331, 40.7395),
    "Queens":    (-73.9626, 40.5431, -73.7004, 40.8007),
    "Bronx":     (-73.9339, 40.7854, -73.7654, 40.9176),
    "Staten Island": (-74.2558, 40.4960, -74.0522, 40.6490)
}

all_areas = []
for name, (lon_min, lat_min, lon_max, lat_max) in bounds.items():
    corners = [
        (lon_min, lat_min),  # Coin Sud-Ouest
        (lon_max, lat_min),  # Coin Sud-Est
        (lon_min, lat_max),  # Coin Nord-Ouest
        (lon_max, lat_max)   # Coin Nord-Est
    ]
    all_areas.append([name, corners])
```

```python
def belongs_to(point, area): 
    # area is a list of 4 tuples for all corners of an area
    # point is a tuple of coordinates to be checked
    
    px, py = point
    
    x_coords = [c[0] for c in area]
    y_coords = [c[1] for c in area]
    
    x_min = area[0][0]
    x_max = area[3][0]
    y_min = area[0][1]
    y_max = area[3][1]
    
   
    if (x_min <= px <= x_max) and (y_min <= py <= y_max):
        return True
    return False
```

```python
# WAY HARDER VERSION, but better:
import geopandas as gpd
from shapely.geometry import Point

url_geojson = "https://data.cityofnewyork.us/api/geospatial/w8k8-3R7t?method=export&format=GeoJSON"
gdf_boroughs = gpd.read_file(url_geojson)
gdf_boroughs = gdf_boroughs[['boro_name', 'geometry']]
gdf_boroughs = gdf_boroughs.set_crs("EPSG:4326") #Code for NYC

def find_areas_geospatial(df, gdf_shapes, lon_col, lat_col, new_area_col): 
    geometry = [Point(xy) for xy in zip(df[lon_col], df[lat_col])]
    gdf_points = gpd.GeoDataFrame(df, geometry=geometry, crs="EPSG:4326") 
    gdf_joined = gpd.sjoin(gdf_points, gdf_shapes, how="left", op="within")
    gdf_joined = gdf_joined.rename(columns={'boro_name': new_area_col}) 
    gdf_joined[new_area_col] = gdf_joined[new_area_col].fillna("Unknown")
    gdf_joined = gdf_joined.drop_duplicates(subset=df.columns)
    
    return gdf_joined

df_with_pickup = find_areas_geospatial(
    df, 
    gdf_boroughs, 
    'pickup_longitude', 
    'pickup_latitude', 
    'Pickup_Area'
)

df_final = find_areas_geospatial(
    df_with_pickup, 
    gdf_boroughs, 
    'dropoff_longitude', 
    'dropoff_latitude', 
    'Dropoff_Area'
)
```

```python
#K-Means


from sklearn.cluster import MiniBatchKMeans


# ... (supposons que 'df' est chargÃ©)

NOMBRE_DE_ZONES = 100

def find_clusters_kmeans(df, lon_col, lat_col, new_cluster_col, n_clusters):
    """
    Applique MiniBatchKMeans pour trouver des clusters.
    """
    print(f"Clustering de '{new_cluster_col}' en {n_clusters} zones...")
    
    # PrÃ©parer les donnÃ©es pour sklearn
    coords_df = df[[lon_col, lat_col]].dropna()
    
    # CHANGEMENT 1: Convertir en array NumPy
    # Cela supprime les noms de colonnes et Ã©vite l'erreur
    coords_np = coords_df.values
    
    kmeans = MiniBatchKMeans(n_clusters=n_clusters,
                             random_state=42,
                             batch_size=2048,
                             n_init='auto')
    
    # EntraÃ®ner le modÃ¨le sur l'array
    kmeans.fit(coords_np)
    
    # PrÃ©dire le cluster pour chaque point
    labels = kmeans.predict(coords_np)
    
    # Assigner les labels au DataFrame (en utilisant l'index de 'coords_df')
    cluster_labels = pd.Series(labels, index=coords_df.index, name=new_cluster_col)
    df_clustered = df.join(cluster_labels)
    df_clustered[new_cluster_col] = df_clustered[new_cluster_col].fillna(-1).astype(int)
    
    print("-> TerminÃ©.")
    return df_clustered, kmeans

# 3. Appliquer pour les DÃ‰PARTS
df_with_pickup, pickup_model = find_clusters_kmeans(
    df, 
    'pickup_longitude', 
    'pickup_latitude', 
    'Pickup_Cluster', 
    n_clusters=NOMBRE_DE_ZONES
)

# 4. Appliquer pour les ARRIVÃ‰ES (Option B CorrigÃ©e)
print("Assignation des zones d'arrivÃ©e (Dropoff)...")

# SÃ©lectionner les donnÃ©es
coords_dropoff_df = df_with_pickup[['dropoff_longitude', 'dropoff_latitude']].dropna()

# CHANGEMENT 2: Convertir les donnÃ©es de dropoff en array NumPy
coords_dropoff_np = coords_dropoff_df.values

# PrÃ©dire sur l'array (le modÃ¨le ne verra jamais les noms "dropoff_")
labels_dropoff = pickup_model.predict(coords_dropoff_np)

# Assigner les labels
dropoff_cluster_labels = pd.Series(labels_dropoff, index=coords_dropoff_df.index, name='Dropoff_Cluster')
df_final = df_with_pickup.join(dropoff_cluster_labels)
df_final['Dropoff_Cluster'] = df_final['Dropoff_Cluster'].fillna(-1).astype(int)

# 5. Analyser les rÃ©sultats
print("\nRÃ©sultat final :")
print(df_final[['Pickup_Cluster', 'Dropoff_Cluster', 'fare_amount']].head())
```

```python
import pandas as pd
from sklearn.cluster import MiniBatchKMeans
import numpy as np

# ... (supposons que 'df' est chargÃ©)

NOMBRE_DE_ZONES = 100

def find_clusters_kmeans(df, lon_col, lat_col, new_cluster_col, n_clusters):
    """
    Applique MiniBatchKMeans de maniÃ¨re Ã©conome en mÃ©moire, en Ã©vitant les .join()
    """
    print(f"Clustering de '{new_cluster_col}' en {n_clusters} zones...")
    
    # --- Partie 1 : PrÃ©parer les donnÃ©es (inchangÃ©e) ---
    coords_df = df[[lon_col, lat_col]].dropna()
    coords_np = coords_df.values
    
    # --- Partie 2 : EntraÃ®ner le modÃ¨le (inchangÃ©e) ---
    kmeans = MiniBatchKMeans(n_clusters=n_clusters,
                             random_state=42,
                             batch_size=2048,
                             n_init='auto')
    
    kmeans.fit(coords_np)
    
    # --- Partie 3 : PrÃ©dire (inchangÃ©e) ---
    labels = kmeans.predict(coords_np)
    
    # --- Partie 4 : Assigner les labels (LA CORRECTION) ---
    
    # Au lieu de :
    # cluster_labels = pd.Series(labels, index=coords_df.index, name=new_cluster_col)
    # df_clustered = df.join(cluster_labels)  <-- CECI CAUSE LA MEMORY ERROR
    # df_clustered[new_cluster_col] = df_clustered[new_cluster_col].fillna(-1).astype(int)

    # NOUS FAISONS CECI (beaucoup plus efficace) :
    print("Assignation des labels (Ã©conome en mÃ©moire)...")
    
    # 1. CrÃ©er un DataFrame de sortie (une copie pour Ãªtre sÃ»r)
    df_clustered = df.copy()
    
    # 2. CrÃ©er la nouvelle colonne, initialisÃ©e Ã  -1 (pour 'noise' ou 'NaN')
    df_clustered[new_cluster_col] = -1
    
    # 3. Utiliser .loc[] pour insÃ©rer les labels UNIQUEMENT aux bonnes lignes
    # C'est l'Ã©tape clÃ©. Elle modifie la colonne sans la recopier.
    df_clustered.loc[coords_df.index, new_cluster_col] = labels
    
    # 4. Assurer le bon type de donnÃ©es
    df_clustered[new_cluster_col] = df_clustered[new_cluster_col].astype(int)
    
    print("-> TerminÃ©.")
    return df_clustered, kmeans

# --- ExÃ©cution (votre code) ---

# 1. Appliquer pour les DÃ‰PARTS
df_with_pickup, pickup_model = find_clusters_kmeans(
    df, 
    'pickup_longitude', 
    'pickup_latitude', 
    'Pickup_Cluster', 
    n_clusters=NOMBRE_DE_ZONES
)

# 2. Appliquer pour les ARRIVÃ‰ES
print("Assignation des zones d'arrivÃ©e (Dropoff)...")
coords_dropoff_df = df_with_pickup[['dropoff_longitude', 'dropoff_latitude']].dropna()
coords_dropoff_np = coords_dropoff_df.values
labels_dropoff = pickup_model.predict(coords_dropoff_np)

# Utilisation de la MÃŠME technique Ã©conome en mÃ©moire pour les dropoffs
df_final = df_with_pickup.copy()
df_final['Dropoff_Cluster'] = -1
df_final.loc[coords_dropoff_df.index, 'Dropoff_Cluster'] = labels_dropoff
df_final['Dropoff_Cluster'] = df_final['Dropoff_Cluster'].astype(int)


# 3. Analyser les rÃ©sultats
print("\nRÃ©sultat final :")
print(df_final[['Pickup_Cluster', 'Dropoff_Cluster', 'fare_amount']].head())
```
